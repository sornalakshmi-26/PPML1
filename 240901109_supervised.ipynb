```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pd.DataFrame({'Age':[36,42,23,52,43,44,66,35,52,35,24,18,45],'Experience':[10,12,4,4,21,14,3,14,13,5,3,3,9],'Rank':[9,4,6,4,8,5,7,9,7,9,5,7,9],'Nationality':['UK','USA','N','USA','USA','UK','N','UK','N','N','USA','UK','UK'],'Go':['NO','NO','NO','NO','YES','NO','YES','YES','YES','YES','NO','YES','YES']})
print(df)

d ={'UK':0,'USA':1,'N':2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES':1,'NO':0}
df['Go']=df['Go'].map(d)

features = ['Age','Experience','Rank','Nationality']
X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()

dtree = dtree.fit(X,y)

tree.plot_tree(dtree, feature_names=features)
plt.show()
```
